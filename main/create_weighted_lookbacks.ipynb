```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bert_2week_ordered.csv')
print(len(df))
df.head
```

```python
#original scores
scores_1 = []
#print(scores_1)
for i in df['bert_scores']:
    scores_1.append(i)
print(scores_1[0:5])
print(len(scores_1))
```

```python
# 1 day past scores
#66-33 weighted scores
#0.66 - 0.33x
scores_2 = [0.166666666]
for i in range(1,len(scores_1)):
    scores_2.append((scores_1[i]*0.66+scores_1[i-1]*0.33)) # x+x/2+x/3 ... to fit only 2 values
print(scores_2[0:5])
print(len(scores_2))
```

```python
#adding 1 day past column
df['2day_weighted'] = scores_2
df.head
```

```python
# 2 day past scores
scores_3 = [0.16666,0.25000]
# 0.45 - 0.1x
for i in range(2,len(scores_1)):
    scores_3.append((scores_1[i]*0.45+scores_1[i-1]*0.35+scores_1[i-2]*0.25)/1.05)
print(scores_3[0:5])
print(len(scores_3))
```

```python
#adding column of 2 days past scores
df['3day_weighted'] = scores_3
#df = df.drop(['Unnamed: 0'], axis = 1)
df.head(8)
```

```python
#df = df.drop(['2week_weighted'], axis = 1)
#df.to_csv('Bert_9year_2-3_windowsize_weighted.csv')
```

```python
# Scores for 7 days
scores_week1 = scores_1[0:6] 
# 0.42 - 0.06x
weight = [0.42,0.36,0.3,0.24,0.18,0.12,0.06]
#weight = [0.25,0.2,0.15,0.15,0.1,0.1,0.5]
for i in range(6,len(scores_1)):
    avg = 0
    for j in range(7):
        avg+=scores_1[i-j]*weight[j]
    scores_week1.append(avg/sum(weight))
print(scores_week1[0:8])
print(len(scores_week1))
#Adding column
df['1week_weighted'] = scores_week1
```

```python
# Scores for 14 days
scores_week2 = scores_1[0:13] 
# 0.26 - 0.02x 
weight = [0.26,0.24,0.22,0.2,0.18,0.16,0.14,0.12,0.1,0.08,0.06,0.04,0.02,0.0]
#weight = [0.15,0.13,0.11,0.1,0.09,0.08,0.07,0.06,0.05,0.04,0.04,0.03,0.3,0.2]
for i in range(13,len(scores_1)):
    avg = 0
    for j in range(14):
        avg+=scores_1[i-j]*weight[j]
    scores_week2.append(avg/sum(weight))
print(scores_week1[0:15])
print(len(scores_week2))
```

```python
#Adding column
df['2week_Weighted'] = scores_week2
```
