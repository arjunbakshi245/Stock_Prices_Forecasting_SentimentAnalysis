```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bert_scores.csv')
print(len(df))
df.head
```

```python
#original scores
scores_1 = []
#print(scores_1)
for i in df['bert_scores']:
    scores_1.append(i)
print(scores_1[0:5])
print(len(scores_1))
```

```python
scores_2 = [0.166666666]
for i in range(1,len(scores_1)):
    scores_2.append((scores_1[i]+scores_1[i-1])/2)
print(scores_2[0:5])
print(len(scores_2))
```

```python
#adding column
df['windowSize_2'] = scores_2
df.head
```

```python
scores_3 = [0.16666,0.25000]
for i in range(2,len(scores_1)):
    scores_3.append((scores_1[i]+scores_1[i-1]+scores_1[i-2])/3)
print(scores_3[0:5])
print(len(scores_3))
```

```python
#adding window of 3 days prior
df['windowSize_3'] = scores_3
df.head(8)
```

```python
df.to_csv('Bert_9year_2-3_windowsize.csv')
```

```python
scores_week1 = scores_1[0:6] 
for i in range(6,len(scores_1)):
    avg = 0
    for j in range(7):
        avg+=scores_1[i-j]
    scores_week1.append(avg/7)
print(scores_week1[0:8])
print(len(scores_week1))
df['week_1'] = scores_week1
```

```python
scores_week2 = scores_1[0:13] 
for i in range(13,len(scores_1)):
    avg = 0
    for j in range(14):
        avg+=scores_1[i-j]
    scores_week2.append(avg/14)
print(scores_week1[0:15])
print(len(scores_week2))
df['week_2'] = scores_week2
```
